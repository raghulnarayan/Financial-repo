```python
import pandas as pd
```

```python
df = pd.read_csv('financials.csv')
```

```python
df
```

```python
# Sort data by Company and Year (just in case)
df = df.sort_values(by=['Company', 'Year'])
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Operating Cash Flow'].pct_change() * 100
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Year'], company_data['Total Revenue'], marker='o', label=company)

plt.title('Total Revenue Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Revenue (in millions)')
plt.legend()
plt.grid(True)
plt.show()
```
